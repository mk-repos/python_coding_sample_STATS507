### Problem 2: Fun with Strings
1. Write a function called is_palindrome, which takes a string as its only argument, and returns a Boolean.

```python
def is_palindrome(s):
    cleaned_s = s.lower().replace(" ", "")
    reversed_s = cleaned_s[::-1]
    if cleaned_s == reversed_s:
        return True
    else:
        return False
```

```python
print(is_palindrome("Was it a car or a cat I saw") == True)
print(is_palindrome("T A C O cat") == True)
```

2. Write a function is_abecedarian, which takes a single argument in the form of a string and returns True if the argument is abecedarian and False otherwise.

```python
def is_abecedarian(s):
    converted_list = list(s.lower().replace(" ", ""))
    sorted_list = sorted(converted_list)
    if converted_list == sorted_list:
        return True
    else:
        return False
```

3. Write a function called double_vowels that takes a string as its only argument and returns that string with all the vowels duplicated.

```python
def double_vowels(s):
    new_string = str()
    for c in s:
        new_string += c
        if c in 'aeiou':
            new_string += c
    return new_string
```

```python
def is_sorted(seq):
    for i in range(len(seq)-1):
        if seq[i] > seq[i+1]:
            return False
    return True
```

3. Write a function called binary_search that takes two arguments, a list of integers t (which is guaranteed to be sorted in ascending order) and an integer elmt, and returns True if elmt appears in list t and False otherwise.

```python
def binary_search(t, elmt):
    if len(t) <= 2:
        for i in range(len(t)):
            if t[i] == elmt:
                return True
        return False
    else:
        median_index = len(t)//2
        if t[median_index] == elmt:
            return True
        elif t[median_index] > elmt:
            return(binary_search(t[:median_index], elmt))
        elif t[median_index] < elmt:
            return(binary_search(t[median_index+1:], elmt))
```

```python
print(binary_search([2,4,6,8,9],9) == True)
print(binary_search([2,4,6,8,9],0) == False)
```

### Problem 4: More Fun with Strings
1. Write a function called char_hist that takes a string as its argument and returns adictionary whose keys are characters and values are the number of times each character appeared in the input.

```python
def char_hist(string):
    lowered_str = string.lower()
    count_dict = dict()
    for char in lowered_str:
        if char in count_dict:
            count_dict[char] += 1
        else:
            count_dict[char] = 1
    return count_dict
```

```python
print(char_hist("gattaca") == {'a':3,'c':1,'g':1,'t':2})
```

2. Write a function called bigram_hist that takes a string as its argument and returns a dictionary whose keys are 2-tuples of characters and values are the number of times that pair of characters appeared in the string.

```python
def bigram_hist(string):
    lowered_str = string.lower()
    count_dict = dict()
    for i in range(len(lowered_str)-1):
        bigram = (lowered_str[i], lowered_str[i+1])
        if bigram in count_dict:
            count_dict[bigram] += 1
        else:
            count_dict[bigram] = 1
    return count_dict
```

```python
print(bigram_hist("mississippi") == {('i','p'):1,('i','s'):2,('m','i'):1,('p','i'):1,('p','p'):1,('s','i'):2,('s','s'):2})
```

### Problem 5: Tuples as Vectors
1. Implement a function called vec_scalar_mult, which takes two arguments: a tuple of numbers (oats and/or integers) t and a number (oat or integer) s and returns a tuple of the same length as t, with its entries equal to the entries of t multiplied by s.

```python
def valid_vector(inputtuple):
    if isinstance(inputtuple, tuple):
        for i in inputtuple:
            if not isinstance(i, int) and not isinstance(i, float):
                return False
        return True
    else:
        return False


def vec_scalar_mult(t, s):
    results_list = []
    if not valid_vector(t):
        raise TypeError('The first argument must be a valid vector')
    elif not isinstance(s, int) and not isinstance(s, float):
        raise TypeError('The second argument must be an int/float')
    else:
        for i in range(len(t)):
            results_list.append(t[i]*s)
    return tuple(results_list)
```

```python
print(vec_scalar_mult((1,2),4) == (4,8))
print(vec_scalar_mult((1,2,3),-2) == (-2,-4,-6))
```

2. Implement a function called vec_inner_product which takes two \vectors" (i.e., tuples of foats and/or ints) as its inputs and outputs a foat corresponding to the inner product of these two vectors.

```python
def vec_inner_product(v1,v2):
    summed = 0
    if not valid_vector(v1) or not valid_vector(v2):
        raise TypeError('The arguments must be valid vectors')
    if not len(v1) == len(v2):
        raise ValueError('The dimensions must be the same')
    else:
        for i in range(len(v1)):
            summed += v1[i]*v2[i]
    return summed
```

```python
print(vec_inner_product((1,1),(1,2)) == 3)
print(vec_inner_product((1,1),(-1,2)) == 1)
```

###  Problem 6: More Fun with Vectors
1. Write a function is_valid_sparse_vector that takes one argument, and returns True if and only if the input is a valid sparse vector, and returns False otherwise.

```python
def is_valid_sparse_vector(d):
    if not isinstance(d, dict):
        return False
    else:
        for k, v in d.items():
            if not isinstance(k, int) or not isinstance(v, float):
                return False
            elif k <= 0:
                return False
        return True
```

```python
print(is_valid_sparse_vector({1:2.,6:1.,5:3.5,7:4.0},) == True)
print(is_valid_sparse_vector({-2:1.2,3:2.5,5:4.1},) == False)
```

2. Write a function sparse_inner_product that takes two \sparse vectors" as its inputs, and returns a foat that is the value of the inner product of the vectors that the inputs represent.

```python
def error_messages_sparse(d):
    if not isinstance(d, dict):
        return TypeError('The argument must be a dictionary')
    else:
        for k, v in d.items():
            if not isinstance(k, int):
                return TypeError('Keys must be a positive integer')
            elif k <= 0:
                return ValueError('Keys must be a positive integer')
            elif not isinstance(v, float):
                return TypeError('Values must be float')


def sparse_inner_product(v1,v2):
    if not is_valid_sparse_vector(v1):
        raise error_messages_sparse(v1)
    elif not is_valid_sparse_vector(v2):
        raise error_messages_sparse(v2)
    else:
        summed = float()
        for k,v in v1.items():
            if k in v2.keys():
                summed += v1[k]*v2[k]
        return summed
```

```python
print(sparse_inner_product({1:1.,2:2.,3:3.},{1:1.,3:8.,7:5.1}) == 25)
```

3. Write a function called rotate_tuple that takes two arguments: a tuple and an integer, in that order.

```python
def rotate_tuple(t,n):
    if not isinstance(t, tuple):
        raise TypeError('The first argument must be a tuple')
    else:
        try:
            axis = abs(n)%len(t)
            if int(n) < 0:
                axis = -axis
            front = t[-axis:]
            rear = t[:-axis]
            return front + rear
        except:
            print('Please provide an integer for the second argument')
```

```python
# Rotate 9 is equal to rotate 2; Similarly, rotate -9 is equal to rotate -2.
print(rotate_tuple((1,2,3,4,5,6,7),9) == (6,7,1,2,3,4,5))
print(rotate_tuple((1,2,3,4,5,6,7),2) == (6,7,1,2,3,4,5))
print(rotate_tuple((1,2,3,4,5,6,7),-9) == (3,4,5,6,7,1,2))
print(rotate_tuple((1,2,3,4,5,6,7),-2) == (3,4,5,6,7,1,2))
```
